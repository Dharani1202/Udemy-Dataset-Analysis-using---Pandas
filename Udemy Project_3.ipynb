```python
import pandas as pd
ds=pd.read_csv('udemy.csv')
ds
```

```python
ds.drop_duplicates()
```

```python
ds.dtypes
```

```python
ds.isnull().sum()
```

```python
#1.	What are all the different subjects for which courses are offered?

ds['subject'].nunique()
```

```python
ds['subject'].unique()
```

```python
#2.	Which subject has maximum number of courses?

ds['subject'].value_counts()
```

```python
#3.	Show all the courses which are free of cost

ds[~ds['is_paid']==True]
```

```python
#4.	Show all the courses which are paid

ds[ds['is_paid']==True]
```

```python
#5.	Which is the top selling course?

ds.sort_values('num_subscribers',ascending=False)
```

```python
ds['num_subscribers'].max()
```

```python
#6.	Which is the least selling course?

ds.sort_values('num_subscribers',ascending=True)
```

```python
#7.	Show all the graphic design course where the price is less than 100
```

```python
a=ds[ds['is_paid']==True]
a
```

```python
ds.dtypes
```

```python
# price in the object form so we need to change the dtype into int 

a['price']=pd.to_numeric(a['price'])
a
```

```python
# now we can check the datatypes    # it converted object into int in the price column
a.dtypes
```

```python
#7.	Show all the graphic design course where the price is less than 100

a[(a['subject']=='Graphic Design')&(a['price']<=100)]
```

```python
#8.	List out all the courses related to python

ds[ds['course_title'].str.contains('Python')]
```

```python
#9.	What are the courses published in the year 2015?

ds['published_timestamp	']=pd.to_datetime(ds.published_timestamp)
ds
```

```python
ds.dtypes
```

```python
ds['Year']=ds['published_timestamp\t'].dt.year
ds
```

```python
# now we can find the courses published in 2015
ds[ds['Year']==2015]
```

```python
#  courses piblished in 2017
ds[ds['Year']==2017]
```

```python
#10. What is the maximum number of subscribers for each level of courses?
ds['level'].value_counts() 
```

```python
ds.groupby('level')['num_subscribers'].max()
```

```python
#11. What is the minimum number of subscribers for each level of courses?

ds.groupby('level')['num_subscribers'].min()
```

```python
# 12. show the web development courses in beginne level

ds[(ds['subject']=='Web Development')&(ds['level']=='Beginner Level')]
```

```python
# 13. show the Musical Instruments courses in expert level

ds[(ds['subject']=='Musical Instruments')&(ds['level']=='Expert Level')]
```
